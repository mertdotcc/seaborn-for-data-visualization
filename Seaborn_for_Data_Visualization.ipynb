### Distribution Plots

```python
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head()
```

Dist plot allows us to show the distribution of a univariate set of observations. __Univariate__ is just a different way of saying _one variable_.

```python
sns.distplot(tips['total_bill'])
```

What we have here is a histogram and that's known as a KDE, kernel density estimation.

We can remove the KDE line by saying that:

```python
sns.distplot(tips['total_bill'], kde=False)
```

We can also change the number of bins, i.e., number of columns.

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

### Joint Plots

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

Jointplot has a parameter called __kind__ and the default value assigned to it is scatter plot, the plot we are seeing right now. We can change the kind parameter and get different looking plots.

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
 sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

### Pairplot

It essentially plots pairwise relationships across an entire data frame at least for the numerical columns.

```python
sns.pairplot(tips)
```

Notice the difference. When we are using __jointplot()__, it basically compares two components of a dataset.

But when we are using __pairplot()__, it pairs up every component of the dataset and makes a pairwise comparison. Note that the bigger the dataset, the more time the function pairplot() is gonna need to compile.

### The hue parameter of pairplot() function.

The pairplot() function comes with a parameter called "hue". What it does is, when we pass a categorical data (categorical data means not a numerical data. A type of data that _categorizes_ the dataset. For example the sex, only two possible values, or the smoker column, either yes or no.) it takes the given category into consideration and creates way more insightful plots.

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips, hue='sex', palette="rainbow")
```

### Rugplots

```python
sns.rugplot(tips['total_bill'])
```

# Matrixplots

```python
import seaborn as sns
import pandas as pd
%matplotlib inline
tips = pd.read_csv('tips.csv')
flights = pd.read_csv('flights.csv')
tips.head()
```

```python
flights.head()
```

### Heatmaps

In order for heatmaps to actually work, our data needs to be in a matrix form.

We need __pivot tables__ or __correlation data__ for our data to be in a matrix form. 

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
import numpy as np
a = np.random.rand(10,10)
b = sns.heatmap(a)
b.figure.savefig("out10.png")
```

```python
a = np.random.rand(100,100)
b = sns.heatmap(a)
b.figure.savefig("out100.png")
```

```python
a = np.random.rand(1000,1000)
b = sns.heatmap(a, cmap='GnBu_r')
b.figure.savefig("out1000b.png", dpi=1000) # You can specify the size of the resolution you want!
```

```python
sns.heatmap(tc, annot=True)
```

```python
flights
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=1)
```

```python
sns.heatmap(fp, cmap='coolwarm', linecolor='black', linewidths=3)
```

### Clustermaps

```python
sns.clustermap(fp, cmap='coolwarm')
```

# Grids

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

iris = pd.read_csv('iris.csv')
iris.head()
```

```python
sns.pairplot(iris)
```

### PairGrid

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

If we want to actually specify what we want in the diagonal, upper half and lower half:

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

### FacetGrid

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```

# Regression Plots

```python
import seaborn as sns
import pandas as pd
%matplotlib inline
tips = pd.read_csv('tips.csv')
tips.head()
```

### Linear Model Plots

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time', scatter_kws={'alpha':0.3, 'color':'green'}, 
                                                                      line_kws={'color':'red'})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex')
```

We can change the __aspect ratio.__

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6, height=8)
```

# Style and Color

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

tips = pd.read_csv('tips.csv')
tips.head()
```

```python
sns.countplot(x='sex', data=tips)
```

### The set_style() argument

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
```

### Removing spines

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
sns.despine()
```

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
sns.despine(left=True, bottom=True)
```

### Size and Aspect

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex', data=tips)
```

### set_context()

```python
sns.set_context('poster')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_context('talk')
sns.countplot(x='sex', data=tips)
```
